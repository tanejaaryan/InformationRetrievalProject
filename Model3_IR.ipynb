```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# url = '../input/coursera-courses-dataset-2021/Coursera.csv'
df = pd.read_csv("/content/gdrive/MyDrive/Coursera.csv")
df.head()
```

```python
uc = ['Course URL','University']  # uc means useless columns
df = df.drop(columns=uc)
df.head()
```

```python
cv=TfidfVectorizer()
tfidf_matrix=cv.fit_transform(df['Course Name'])
course_user = df.pivot_table(columns='Course Name',values='Course Rating')
course_user.head()

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
indices=pd.Series(df.index,index=df['Course Name'])
titles=df['Course Name']
def recommendations(course_title):
    idx = indices[course_title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    course_indices = [i[0] for i in sim_scores]
    return titles.iloc[course_indices]
print("--------------- Similar courses to your search --------------:\n")
recommendations('Write A Feature Length Screenplay For Film Or Television')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import regex as re
from nltk.stem import PorterStemmer
from wordcloud import STOPWORDS
import numpy as np
import nltk
nltk.download("stopwords")
stops = set(stopwords.words('english'))
for i in stops:
    STOPWORDS.add(i)

ps = PorterStemmer()
nltk.download('punkt')

def  processString(text):
  # print("___")
  symbols = "!\"#$%&()*+-./:;,<=>?@[\]^_`{|}~\n"
  for num in symbols:
    text = text.replace(num, " ")
    text = text.replace("  ", " ")
  text = text.replace("'", " ")
  text = text.replace("  ", " ")
  tokens_in_text = word_tokenize(text)
  for i in range(len(tokens_in_text)):
    token = tokens_in_text[i]
    token = token.strip()
    token = token.lower()
    token = ps.stem(token)
    tokens_in_text[i] = token;
  tokens_in_text =  [w for w in tokens_in_text if w not in STOPWORDS]
  return tokens_in_text;


df["CourseDescription_tokens"] = df["Course Description"].apply(lambda x: x.split())

```

```python
from gensim.models.word2vec import Word2Vec

word2vec_model = Word2Vec(min_count=0, workers = 8, size=300) 
# Prepare vocab
word2vec_model.build_vocab(df.CourseDescription_tokens.values)
# Train
word2vec_model.train(df.CourseDescription_tokens, total_examples=word2vec_model.corpus_count, epochs=30)

```

```python
# Predict
def is_word_in_model(word, model):
    print(model)
    is_in_vocab = word in model.vocab
    return is_in_vocab

def predict(query_sentence, dataset, model, topk=3):
    query_sentence = query_sentence.split()
    in_vocab_list, best_index = [], [0]*topk
    for w in query_sentence:
        # remove unseen words from query sentence
        if is_word_in_model(w, model.wv):
            in_vocab_list.append(w)
    # Retrieve the similarity between two words as a distance
    if len(in_vocab_list) > 0:
      
        sim_mat = np.zeros(len(dataset))  # TO DO
        print(sim_mat)
        for i, data_sentence in enumerate(dataset):
            if data_sentence:
                sim_sentence = model.wv.n_similarity(
                        in_vocab_list, data_sentence)
            else:
                sim_sentence = 0
            sim_mat[i] = np.array(sim_sentence)
        # Take the five highest norm
        best_index = np.argsort(sim_mat)[::-10][:topk]
    return best_index

best_index = predict("Java", df['CourseDescription_tokens'].values, word2vec_model)

df.iloc[best_index]    
# display(df[['original_title', 'genres', 'sentence']].iloc[best_index])

```
